```python
"""Q6. List down some best practices for exception handling in python."""

""""1) Avoid raising generic exceptions because if we do so, then all other more 
      specific exceptions have to be caught also. Hence, the best practice is to raise the most 
      specific exception close to our problem."""

        

```

```python
""""1) Avoid raising generic exceptions because if we do so, then all other more 
      specific exceptions have to be caught also. Hence, the best practice is to raise the most 
      specific exception close to our problem."""
try:
    4/0
except ZeroDivisionError as e:
    print(e)
```

```python
"""2) Always try to print a proper message so that it is become readile and understable for the programmer that 
      what the specific code is written for.

"""

try:
    4/0
    
except ZeroDivisionError as e:
    print('i have found',e)
```

```python
"""3) Always try to log the answer instead of print function because logging helps to find that where error has occured
      and in future we can easily debug that error and can find what is the problem in the code and we can easily use that 
      for our future use."""

import logging
logging.basicConfig(filename='testt.log',level=logging.ERROR)

try:
    2/0
except ZeroDivisionError as e:
    logging.error("I have an zero divison error {}".format(e))
```

```python
##4) Always try to avoid multiple exception handling

try:
    2/0
    
except FileNotFoundError as e:
    logging.error('i am handling file not found {}'.format(e))
    
    
except AttributeError as e:
    logging.error('i am handling Attribute error {}'.format(e))
        
except IndexError as e:
    logging.error('i am handling index error {}'.format(e))
    
except ZeroDivisionError as e:
    logging.error('i am handling ZeroDivisonError {}'.format(e))
        
    
```
